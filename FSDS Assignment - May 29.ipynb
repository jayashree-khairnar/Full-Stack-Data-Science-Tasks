#### 15. Write a fun which will be able to print an index of all primitive element which you will pass

```python
def index_search(string, to_find_indx_of):
    
    lst= list(string.split())
    word_len= 0
    
    for i in lst:
        if i != to_find_indx_of:
            word_len += len(i) + 1
            
        if i == to_find_indx_of:
            return word_len
        
    else:
        return False
        
string= str(input("Say something... "))
to_find_indx_of= str(input("Now enter a string from above sentence to get its index: "))

index= index_search(string, to_find_indx_of)

if index != False:
    print("\nIndex of '{}': {}".format(to_find_indx_of, index))
    
else:
    print("\n'{}' word is not present.".format(to_find_indx_of))
```

#### 16. Write a fun which will take input as a dict and give output as a list of all the values even in case of 2 level nesting it should work

```python
def dict_to_list():

    dictionary= {'k1':'val1', 'k2':{1:'one', 2:'two'}, 'k3':{3:'three', 4:'four'}}

    lst= []

    for i,j in dictionary.items():
        lst.append(i)

        if type(j) == dict:
            for p,q in j.items():
                lst.append(p)
                lst.append(q)

        elif type(j)==str or type(j)==int:
            lst.append(j)
    
    return lst

dict_to_list()
```

#### 20. Write a function by which you will be able to append two PDF files

```python
from PyPDF2 import PdfMerger

write_pdf_to_merge = ['LexA.pdf', 'RE.pdf']

merger = PdfMerger()

for pdf in write_pdf_to_merge:
    merger.append(pdf)

merger.write("merged_LexA_RE.pdf")
merger.close()

for i in os.listdir():
    if i == 'merged_LexA_RE.pdf':
        print(i)
```

#### 22. Write a function which can read video file and play for you

```python
import cv2

cap= cv2.VideoCapture("video.mp4")

while(cap.isOpened()):
    ret, frame= cap.read()
    
    frame= cv2.resize(frame,(300,700))
    cv2.imshow("video", frame)
   
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
